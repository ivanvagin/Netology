### Домашнее задание №4 - Функции нескольких переменных
Задание 1.
Постройте трехмерный график функции  
$f(x)=x_2^2-x_1^2 x_2 + 2 x_1 x_2$  
А также график линий уровня с уровнями 1,5,10

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = Axes3D(fig)
X = np.arange(-2,2,0.1)
Y = np.arange(-2,2,0.1)
X,Y = np.meshgrid(X,Y)

def f(x2,x1):
    return (x2**2-x1**2 * x2 + 2*x1*x2)

ax.plot_surface(X,Y,f(X,Y), rstride=1, cstride=1)

# Customize the z axis.
ax.set_zlim(-10, 10.0)
ax.zaxis.set_major_locator(LinearLocator(10))
ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))

# Add a color bar which maps values to colors.
# fig.colorbar(surf, shrink=0.5, aspect=5)

ax.azim = 50
ax.dist = 8
ax.elev = 12

levels = [1,5,10]
plt.contour(X, Y, Z, levels=levels)
plt.show()
```

### Задание 2.
Вычислить вектор градиента и матрицу Гессе для функции:   

$f(x)=2x_1^3-3x_1 x_2+2x_2 x_3^2-x_2^3$ 
    
    
   в точке $M=(2,1,1)$





```python
import sympy as sym
from sympy import *
```

```python
x1 = sym.Symbol('x1')
x2 = sym.Symbol('x2')
x3 = sym.Symbol('x3')
```

Находим вектор градиента

```python
f = 2*x1**3-3*x1*x2+2*x2*x3**2-x2**3
```

```python
fx1=f.diff(x1)
print('df/dx =',fx1,'=0')
fx2=f.diff(x2)
print('df/dx3 =',fx2,'=0')
fx3=f.diff(x3)
print('df/dz =',fx3,'=0')
```

```python
sym.diff(2*x1**3-3*x1*x2+2*x2*x3**2-x2**3,x1)
```

```python
sym.diff(2*x1**3-3*x1*x2+2*x2*x3**2-x2**3,x2)
```

```python
sym.diff(2*x1**3-3*x1*x2+2*x2*x3**2-x2**3,x3)
```

Находим матрицу Гессе 

Находим вторые частные производные в точке

```python
fx1x1=f.diff(x1,x1).subs({x1:2,x2:1,x3:1})
print('fx1x1=',fx1x1)
fx2x2=f.diff(x1,x2).subs({x1:2,x2:1,x3:1})
print('fx1x2=',fx1x2)
fx3x3=f.diff(x1,x3).subs({x1:2,x2:1,x3:1})
print('fx1x3=',fx1x3)
fx3x3=f.diff(x2,x2).subs({x1:2,x2:1,x3:1})
print('fx1x3=',fx2x2)
fx3x3=f.diff(x3,x2).subs({x1:2,x2:1,x3:1})
print('fx1x3=',fx3x2)
fx3x3=f.diff(x3,x3).subs({x1:2,x2:1,x3:1})
print('fx1x3=',fx3x3)
```

```python
d1=fxx
print('Определитель М1- d1=',d1)
M2=Matrix([[fx1x1,fx1x2],[fx2x1,fx2x2]])
print('Матрица М2-',M2)
d2=M2.det()
print('Определитель М2- d2=',d2)
M3=Matrix([[fx1x1,fx1x2,fx1x3],[fx2x1,fx2x2,fx2x3],[fx3x1,fx3x2,fx3x3]])
print('Матрица М3-',M3)
d3=M3.det()
print('Определитель М3- d3=',d3)
```
