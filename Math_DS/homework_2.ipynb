### Домашнее задание №2 - Продвинутая линейная алгебра
Задание 1.
Дан объект в  2D пространстве. При помощи линейных отображений:

Уменьшить объект в два раза  
Повернуть на 130 градусов  
Отразить объект относительно прямой y=x  
Затем вычислить определитель композиции преобразований (итоговой матрицы реализующей все три отображения)  

```python
import numpy as np
import matplotlib.pyplot as plt
import math as m

A = [[-100, -100, 1],
[-100, 100, 1],
[100, 100, 1],
[0, 0, 1],
[100, -100, 1],
[-100, -100, 1]]

A = np.array(A)
x = A[:,0]
y = A[:,1]

plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

Уменьшить объект в два раза

```python
B = np.array([[0.5,0,0], [0,0.5,0], [0,0,0.5]])
res=B@A.transpose()
decrease = res.transpose()
decrease
```

```python
A = np.array(decrease)
x = A[:,0]
y = A[:,1]

plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

Повернуть на 130 градусов

```python
A1 = [[-100, -100, 1],
[-100, 100, 1],
[100, 100, 1],
[0, 0, 1],
[100, -100, 1],
[-100, -100, 1]]

A1 = np.array(A1)
x = A1[:,0]
y = A1[:,1]

plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

```python
angle = np.radians(130)
T = np.float32([ [m.cos(angle), -(m.sin(angle)), 0],
              [m.sin(angle), m.cos(angle), 0],
              [0, 0, 1] ])
res1=T@A.transpose()
turn = res1.transpose()
turn
```

```python
A1 = np.array(turn)
x = A1[:,0]
y = A1[:,1]

plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

Отразить объект относительно прямой y=x

```python
A2 = [[-100, -100, 1],
[-100, 100, 1],
[100, 100, 1],
[0, 0, 1],
[100, -100, 1],
[-100, -100, 1]]

A2 = np.array(A2)
x = A2[:,0]
y = A2[:,1]

plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

```python
R = np.array([[ 0, 1, 0],
              [1,0, 0],
              [0, 0, 0] ])
res2=R@A2.transpose()
reflection = res2.transpose()
reflection
```

```python
A2 = np.array(reflection)
x = A2[:,0]
y = A2[:,1]

plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

Вычислить определитель композиции преобразований (итоговой матрицы реализующей все три отображения)

```python
total = (B@T@R)
```

```python
total
```

```python
np.linalg.det(total)
```

### Задание 3.
Найдите спектральное разложение матрицы

```python
m = np.array([[1, 2],
              [2,3]])
```

```python
# находим собственные значения и собвственные вектора нашей матрицы
v,w = np.linalg.eigh(m)
# собственные значение
v
```

```python
# собственные вектора
w
```

```python
# создадим диагональную матрицу из собственных значений
d = np.diag(v)
d
```

```python
# создаем две ортогональные матрицы из собственных векторов
S = w.T
S
```

```python
S.T
```

```python
# проверим что полученные матрицы ортогональные . Для этого возьмем их произведениеи
np.dot(S.T,S)
```

```python
# проверим. что полученные матрицы образуют спектральное разложение 
np.dot(np.dot(S.T,d),S)
```
